```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.model_selection import train_test_split, GridSearchCV
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.svm import LinearSVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import Lasso, LassoCV
```

```python
# Below, set your own path where you have stored the data file. 
housing = pd.read_csv("housing.csv")
```

```python
raw_df = pd.DataFrame(housing)
```

```python
raw_df.info()
```

```python
df = raw_df.dropna()
df.info()
```

```python
less_than_500001 = df[df["median_house_value"] <= 500000]
# Check so that it worked
less_than_500001[less_than_500001["median_house_value"] > 500000]
```

```python
df = less_than_500001
df
```

```python
df = df.join(pd.get_dummies(df.ocean_proximity)).drop("ocean_proximity", axis=1)
df
```

```python
X = df.drop("median_house_value", axis=1)
y = df["median_house_value"]
```

```python
# X_train = X[:5000]
# y_train = y[:5000]
# X_val = X[5000:6000]
# y_val = y[5000:6000]

# X_test = X[6000:7000]
# y_test = y[6000:7000]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Linear Regression

```python
regr = make_pipeline(StandardScaler(), LinearSVR(random_state=0, tol=1e-5))
```

```python
param_grid = {
    'linearsvr__C': [0.1, 1, 10, 100],
    'linearsvr__epsilon': [0.01, 0.1, 1],
    'linearsvr__max_iter': [500000],  # Increase max_iter
    'linearsvr__tol': [1e-4, 1e-3, 1e-2]
}

grid_search = GridSearchCV(regr, param_grid)
```

```python
grid_search.fit(X_train_scaled, y_train)
```

```python
y_pred = grid_search.predict(X_test_scaled)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"LinearSVR MSE: {mse}")
```

```python
r2 = r2_score(y_test, y_pred)

print(f'R^2 Score: {r2}')
```

# Lasso Regression

```python
lasso_cv = LassoCV(cv=5, random_state=42)
lasso_cv.fit(X_train_scaled, y_train)
```

```python
best_alpha = lasso_cv.alpha_
print(f"Best alpha: {best_alpha}")
```

```python
final_lasso_model = Lasso(alpha=best_alpha, random_state=42)
final_lasso_model.fit(X_train_scaled, y_train)
```

```python
predictions = final_lasso_model.predict(X_test_scaled)
```

```python
mse = mean_squared_error(y_test, predictions)
print(f"Final Lasso Regression MSE: {mse}")
```

```python
r2 = r2_score(y_test, predictions)
print(f'R^2 Score: {r2}')
```
